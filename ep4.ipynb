Taxa de mortalidade feminina por câncer de 1989 até 2016

```python
ajeitado = pd.read_csv(r'C:\Users\User\Desktop\Fatec Robson\EstruturaDados\Feminino.csv', sep=';', header=None, names = cabecalho, index_col = 0)
print(ajeitado)
```

Câncer de Mama

```python
plt.figure(figsize=(10,7))
plt.plot(ajeitado.Mama)
plt.xlabel('Ano')
plt.ylabel('Percentual')
plt.title('Taxa de Mortalidade Feminina\nCâncer de Mama')
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.plot(ajeitado.Colon)
plt.xlabel('Ano')
plt.ylabel('Percentual')
plt.title('Taxa de Mortalidade Feminina\nCâncer de Colon')
plt.show()
```

```python
plt.figure(figsize=(17,25))
plt.plot(ajeitado.Bronquios_e_Pulmões, label = 'Brônquios e Pulmões')
plt.plot(ajeitado.Colo_do_Utero, label = 'Colo do´Útero ')
plt.plot(ajeitado.Estômago, label = 'Estômago')
plt.plot(ajeitado.Colon, label = 'Colon')
plt.plot(ajeitado.Mama, label = 'Mama')
plt.title('Taxa de Mortalidade Feminina\nCâncer')
plt.xlabel('Ano')
plt.ylabel('Percentual')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
